```python
import pandas as pd
```

```python
dict =  {"a" : [1, 4, 7],"b" :[2, 5, 8],"c" : [3,6,9]}
df1  = pd.DataFrame(dict)
```

```python
df1
```

```python
dict ={"b" : [3, 6, 9],"c" : [4, 7, 10], "d" :[5, 8, 11]} 
df2  = pd.DataFrame(dict)

df2
```

```python
 df1.info()       # checking the info
```

```python
df1['a'].value_counts()
(df1)
(df2)  # check the unique value count in a particular column
```

```python
# concatenate two data frame and create a third data frame
#concatenate based on y axis, default its an outer join
df3 = pd.concat([df1,df2],ignore_index=True)
df3
```

```python
df5 = pd.concat([df1,df2], join = 'inner', ignore_index=True)
df5
```

```python
df4 = pd.concat([df1,df2],axis = 1 , ignore_index=True)
df4
```

```python
df6 = pd.concat([df1,df2],axis = 1, join = 'inner', ignore_index=False)
df6
```

```python
#handle null values
df3
```

```python
df3.isna()
```

```python
df3.isnull().sum()
```

```python
#drop columns which has missing values
df3.dropna(axis= 1)
```

```python
#drop rows which has missing values
df3.dropna()
```

```python
#Drop the rows where all elements are missing.
df3.dropna(how='all')
```

```python
# Drop rows based on the threshold
df3.dropna(thresh=2, axis = 1)
```

```python
df3.dropna(thresh=4, axis = 1)
```

```python
#Define in which columns to look for missing values.
df3.dropna(subset=['a'])
```

```python
#Fill missing values
df3.fillna(0)
```

```python
#Only replace the first NaN element.
df3.fillna(value=0, limit=1)
```

```python
#Group By in pandas
df3.groupby(['b']).mean()
```
